# 《NBA球队数据爬取器》



```python
from bs4 import BeautifulSoup  # 网页解析，获取数据
import re  # 正则表达式，进行文字匹配
import urllib.request, urllib.error  # 数据
import numpy as np
import pandas as pd
# import seaborn as sns; 
%matplotlib inline
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')          #复制模块
import xlwt                                               #导入excel模块
from pandas.core.frame import DataFrame                       #导入dataframe模块
```

```python
#正则表达
findlinkandname=re.compile(r'<b><a href="(.*?)" target="_blank">(.*?)</a></b>')
findimg=re.compile(r'<img.*src="(.*?)"',re.S)                     #球员图片
findnum = re.compile(r'<td>(\d{1,2})</td>')                       #球员号码
findposition = re.compile(r'<td>([A-Z-]+)</td>')                  #球员位置
findheight = re.compile(r'<td>(.*?米/.*?)</td>')                  #球员身高
findweight = re.compile(r'<td>(.*?公斤/.*?)</td>')                 #球员体重
findbirth = re.compile(r'<td>(\d{4}-\d{1,2}-\d{1,2})</td>')       #球员出生日期
findsalary = re.compile(r'<b>(本年薪金\：.*)</b>')                  #球员薪资
```

# 函数

```python
def askurl(url):                           #爬取数据
# 模拟浏览器头部信息，向服务器发送消息
    head = {
        ""
    }  # 用户代理
    request = urllib.request.Request(url, headers=head)
    html = ""
    try:
        response = urllib.request.urlopen(request)
        html = response.read().decode("utf-8")
        print(html)                                   #测试
    except urllib.error.URLError as e:
        if hasattr(e, "code"):
            print(e.code)
        if hasattr(e, "reason"):
            print(e.reason)
    return html
```

```python
def datafilter(url):
    datalist=[]
    html=askurl(url)
    soup=BeautifulSoup(html,"html.parser")
    for item in soup.find_all("tr"):
        #print(item)           #测试：查看球员信息位置
        data=[]
        item=str(item)
        #球员链接
        #link=re.findall(findlink,item)
        #print(link)           #测试：查看球员链接    
       # data.append(link)
        #name=re.findall(findname,item)
        #data.append(name)
        linkandname=re.findall(findlinkandname,item)
        data.append(linkandname)
        img=re.findall(findimg,item)
        data.append(img)
        num=re.findall(findnum,item)
        data.append(num)
        position=re.findall(findposition,item)
        data.append(position)
        height=re.findall(findheight,item)
        data.append(height)
        weight=re.findall(findweight,item)
        data.append(weight)
        birth=re.findall(findbirth,item)
        data.append(birth)
        salary=re.findall(findsalary,item)
        if len(salary):
            salary=salary[0]
            salary=salary.strip('本年薪金：').strip('万美元')
            if salary=='':
                salary = 0
        data.append('本年薪金：' + str(salary) + '万美元')                        
        datalist.append(data)
    #print(datalist)                     #测试输出
    data1=DataFrame(datalist)
    data1=data1.drop(index=0)
    return data1
```

# 查询程序本体

```python

url_head="https://nba.hupu.com/players/"
teamslist = ['grizzlies','mavericks','spurs','rockets','pelicans','warriors','suns','clippers','lakers',
     'jazz','nuggets','timberwolves','blazers','thunder','nets','76ers','celtics','raptors','knicks',
     'heat','wizards','hornets','hawks','magic','bulls','bucks','cavaliers','pacers','pistons']
whether=['是','好的','好','yes','Yes','y']
i=input("请输入要查询球队的英文名")
if i in teamslist:
    url=(url_head+i)
    print(datafilter(url) )
    data=datafilter(url)
    print("加载完毕")
    s=input("是否保存至本地")
    if s in whether:
        data.to_excel(i+'.xls',sheet_name='Teaminformation')
        print("保存完毕")
    else:
        print("好的")
else:
    print("球队名称错误？")
```
